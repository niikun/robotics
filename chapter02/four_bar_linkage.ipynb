```python
import numpy as np
import matplotlib.pyplot as plt

# 定義: 4バーリンクの仮想的な1次元曲線を可視化（概念的なデモ）

# サンプルデータ生成（θ1: 入力角度、θ2, θ3, θ4は制約に基づく値）
theta1 = np.linspace(0, 2 * np.pi, 100)  # θ1: 自由に動ける入力角度
theta2 = np.sin(theta1)
theta3 = np.cos(theta1)
theta4 = 2*np.pi - theta1 - theta2 - theta3

plt.figure(figsize=(8,6))
plt.plot(theta1,theta2, label="Projected 1D Curve (θ1 vs θ2)", color="blue")
plt.plot(theta1,theta3, label="Projected 1D Curve (θ1 vs θ3)", color="red")
plt.plot(theta1,theta4, label="Projected 1D Curve (θ1 vs θ4)", color="yellow")
plt.title("1D Curve in 4D Joint Space (Projection onto θ1 vs others)", fontsize=14)
plt.xlabel("θ1 (rad)", fontsize=12)
plt.ylabel("θ2 (rad)", fontsize=12)
plt.axhline(0, color="gray", linestyle="--", linewidth=0.5)
plt.axvline(0, color="gray", linestyle="--", linewidth=0.5)
plt.legend()
plt.grid()
plt.show()
```
